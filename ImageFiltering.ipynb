## Image Filtering

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('data/city.jpg',0)

laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
blur = cv2.blur(img,(5,5))
gaussianBlur = cv2.GaussianBlur(img,(5,5),0)
```

```python
plt.imshow(img,cmap = 'gray')
plt.title('Original')
plt.show()
plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian')
plt.show()
plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X')
plt.show()
plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y')
plt.show()
plt.imshow(blur,cmap = 'gray')
plt.title('Averaging')
plt.show()
plt.imshow(gaussianBlur,cmap = 'gray')
plt.title('Gaussian')
plt.show()
```

### **1. Laplacian Filter**

**Effect:** Highlights **regions of rapid intensity change**  i.e., **edges**.

* Responds to **edges in all directions**.

---

### **2. Sobel Filter**

**Effect:** Detects **edges**, but with **directional sensitivity** (horizontal/vertical).

* Commonly applied as **Sobel X** and **Sobel Y** to detect edges in those directions.

---

### **3. Averaging Filter (Box Filter)**

**Effect:** **Smooths** or **blurs** the image.

* Each pixel becomes the **average** of its neighbors.
* Simple and fast, but may blur edges too much.

---

###  **4. Gaussian Filter**

**Effect:** **Smooths** the image while **preserving structure** better than averaging.

* Uses a **weighted average** (Gaussian kernel)  closer pixels matter more.
* Better at reducing **Gaussian noise** with minimal edge distortion.
